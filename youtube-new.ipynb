```python
import csv
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
```

```python
table = pd.read_csv(r'CAvideos.csv')
print(table)
```

# Ê†áÁß∞Â±ûÊÄß

```python
NominalAttribute = ['trending_date', 'publish_time', 'category_id','comments_disabled','ratings_disabled','video_error_or_removed']
```

ËæìÂá∫ÂêÑ‰∏™Â±ûÊÄßÁöÑÈ¢ëÊï∞

```python
for i in NominalAttribute:
    attrValue = []
    for j in table[i]:
        if j==j:
            attrValue.append(j)
    count = Counter(attrValue)
    count_dict = dict(count)
    print('{}ÁöÑÈ¢ëÊï∞Ôºö'.format(i))
    print(count_dict)
```

# Êï∞ÂÄºÂ±ûÊÄß
1. views

```python
#viewsÁöÑ‰∫îÊï∞Ê¶ÇÊã¨
views_r = []
views = []
for i in table['views']:
    views_r.append(i)
    if i == i:
        views.append(i)
miss = 0
print('ÊúÄÂ§ßÂÄºÔºö', max(views))
print('ÊúÄÂ∞èÂÄºÔºö', min(views))
print('‰∏≠‰ΩçÊï∞Ôºö', np.median(views))
print('ÂõõÂàÜ‰ΩçÊï∞Ôºö', np.percentile(views, (25, 50, 75), interpolation='midpoint'))
for i in views_r:
    if i != i:
        miss += 1
print('Áº∫Â§±ÂÄº‰∏™Êï∞Ôºö', miss)
```

```python
#viewsÁöÑÁõ¥ÊñπÂõæ
plt.hist(views_r, 2000)
plt.xlabel('views')
plt.ylabel('number')
plt.xlim(min(views), 3000000)
plt.ylim(0.0, 5000)
plt.title('views')
plt.show()
```

```python
#viewsÁõíÂõæ
data = pd.DataFrame({'views': views})
data.boxplot()
plt.ylabel('views')
plt.xlabel('')
plt.show()
```
